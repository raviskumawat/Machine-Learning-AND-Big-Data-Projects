```python
import matplotlib.pyplot as plt
import scipy.io.wavfile as wav
from scipy import signal
import IPython.display as play
import numpy as np
%matplotlib inline

sample_rate,samples=wav.read('audio.wav')
freq,times,spectrogram=signal.spectrogram(samples,sample_rate)


plt.subplot(211)

plt.pcolormesh(times, freq,np.log(spectrogram))
#plt.imshow(np.log(spectrogram))

plt.subplot(212)

plt.pcolormesh(times, freq,10*np.log(spectrogram+1e-15))

plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()



```
